```python
age_dummies = pd.get_dummies(data['Age'])
```

```python
reason_ab = pd.get_dummies(data["Reason for Absence"],drop_first = True)##Drop reason 0 because it stands for 'Unkown' reason
```

```python
data.drop(columns = ['Reason for Absence'], inplace = True)
```

```python
reason_ab.loc[:,'1':'14'].head()
```

### We will group the reasons for absence into 4 groups depending on the feature description:

```python
reason1 = reason_ab.loc[:, 1:14].max(axis = 1)
reason2 = reason_ab.loc[:, 15:17].max(axis = 1)
reason3 = reason_ab.loc[:, 18:21].max(axis = 1)
reason4 = reason_ab.loc[:, 22:].max(axis = 1)
```

```python
reason1[:5]
```

```python
data = pd.concat([data, reason1, reason2, reason3, reason4], axis = 1)
```

```python
data.head()
```

```python
data.rename(columns = {0:"R1", 1:"R2", 2:"R3", 3:"R4"}, inplace = True)
```

### Extract month value:

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months[5:10]
```

```python
len(list_months)
```

```python
df_reason_mod['Month_value'] = list_months
```

```python
df_reason_mod.head()
```

### Extract the Day of the week

```python
df_reason_mod['Date'][3].day
```

```python
df_reason_mod['Date'][3].weekday()
```

```python
df_reason_mod['weekday'] = df_reason_mod['Date'].apply(lambda date: date.weekday())
```

```python
df_reason_mod.drop(columns = ['Date'], inplace = True)
```
